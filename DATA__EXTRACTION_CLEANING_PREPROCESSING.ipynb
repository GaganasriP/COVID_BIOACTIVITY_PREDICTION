**DATA EXTRACTING, CLEANING, PREPROCESSING**

The data was extracted from ChEMBL site focusing only on the SARS coronavirus. It was filtered and consolidated to contain standard value(bioactivity value) and the canonical smiles on the basis of target type- single protein and standard type-IC50.
Then using rdkit, the final data was generated from the respective descriptors using the canonical smiles.

After more data cleaning and avoiding compounds with molecular weight>=500, DATA.csv file is obtained for modelling.

```python
NEW=DF.dropna()
NEW
```

```python
import numpy as np
from rdkit import Chem
from rdkit.Chem import Descriptors
```

```python
def discriptor(smiles, verbose=False):
    moldata= []
    for elem in smiles:
        mol=Chem.MolFromSmiles(elem)
        moldata.append(mol)
    baseData= np.arange(1,1)
    i=0
    for mol in moldata:
        desc_MolWt = Descriptors.MolWt(mol)
        desc_PSA = Descriptors.TPSA(mol)
        desc_MolLogP = Descriptors.MolLogP(mol)
        desc_MolHBA = Descriptors.NumHAcceptors(mol)
        desc_MolHBD = Descriptors.NumHDonors(mol)

        row = np.array([desc_MolWt,
                        desc_PSA,
                        desc_MolLogP,
                        desc_MolHBA,
                        desc_MolHBD])
        if(i==0):
            baseData=row
        else:
            baseData=np.vstack([baseData, row])
        i=i+1

    columnNames=["MOLECULAR WEIGHT","PSA","logP","HBA","HBD"]
    descriptors = pd.DataFrame(data=baseData,columns=columnNames)

    return descriptors
```

```python
DATADISCRIPT = discriptor(NEW.canonical_smiles)
DATADISCRIPT
```

```python
DATA_NEW = pd.concat([NEW,DATADISCRIPT],axis=1)
DATA_NEW
```

```python
DF=DATA_NEW.dropna()
DF
```

```python
DF=DF[DF['MOLECULAR WEIGHT'] <= 500]
DF
```

```python
DF.to_csv('DATA.csv', index=False)
```
